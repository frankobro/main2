```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('james_bond_data_cleansed.csv')
data.head(10)
```

```python
import numpy as np
```

```python
x = np.array(data['imdb'])
y = np.array(data['rotten_tomatoes'])


plt.xlabel('Avg IMDb Rating')
plt.ylabel('Avg Rotten Tomatoes Rating')
plt.title('Ratings Plot')
plt.savefig('Ratings Plot.jpg')


plt.scatter(x, y)
plt.show()
```

```python
x = np.array(data['imdb'])
y = np.array(data['bond_kills'])


plt.xlabel('Avg IMDb Rating')
plt.ylabel('James Bond Kills')
plt.title('James Bond Kills vs IMDb Rating')
plt.savefig('James Bond Kills vs IMDb Rating.jpg')


plt.scatter(x, y)
plt.show()
```

Answer: There is no correlation between the number of bad guys killed by bond and the IMDb Rating.

```python
x = data['release_year']
y = data['income_world']
y1 = data['movie_budget']

plt.plot(x, y1, 'b-o')
plt.plot(x, y, 'r-o')
plt.xlabel('Release Year')
plt.legend()
plt.show()
```

```python
x = data['release_year']
y = data['income_world'] / data['movie_budget']  

# Plot the data
plt.bar(x, y, label='Income/Budget Ratio')
plt.xlabel('Release Year')
plt.ylabel('Income/Budget Ratio')
plt.title('Income to Budget Ratio by Release Year')
plt.legend()
plt.show()
```
